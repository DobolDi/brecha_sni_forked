```python
f = "../data/002_processed/sni.zip"
tmp = pd.read_csv(f, compression="zip")
```

```python
# SNI solo tiene los campos definidos con la lista grados,
# pues son los únicos que se pueden identificar como hombre,mujer
grados = ["DR.","DRA.","MTRO.","MTRA."]
sni = tmp[tmp["grado"].isin(grados)].copy()
mapeo_sexo = {
    'DR.': 'H',
    'MTRO.': 'H',
    'DRA.': 'M',
    'MTRA.': 'M'
}

# Crear la nueva columna "sexo" usando el mapeo
sni['sexo'] = sni['grado'].map(mapeo_sexo)
```

```python
sni.anio.unique()
```

```python
historico = sni.groupby(["anio"]).size()
historico
```

```python
total_anios = sni.groupby(["anio", "sexo"]).size().unstack(fill_value=0)
#Moví esta celda acá porque originalmente estaba abajo de la gráfica y la gráfica no corría.
```

```python
fig, ax = plt.subplots(figsize=(10,6))


# Eliminamos el eje y (ticks, labels y línea del eje)
ax.yaxis.set_visible(False)
ax.tick_params(axis='y', which='both', left=False, labelleft=False)


# --- Trazamos las líneas para Hombres y Mujeres ---
ax.plot(total_anios.index, total_anios["H"], color="orange", label="Hombres", linewidth=2)
ax.plot(total_anios.index, total_anios["M"], color="seagreen", label="Mujeres", linewidth=2)

# --- Relleno del área entre las dos curvas para resaltar la diferencia ---
ax.fill_between(total_anios.index, total_anios["H"], total_anios["M"],
                where=(total_anios["H"] >= total_anios["M"]),
                interpolate=True, color="orange", alpha=0.3, label="Mayor Hombres")
ax.fill_between(total_anios.index, total_anios["H"], total_anios["M"],
                where=(total_anios["H"] < total_anios["M"]),
                interpolate=True, color="seagreen", alpha=0.3, label="Mayor Mujeres")

# --- Formateo de ejes ---
ax.set_xlabel("Año")
ax.set_ylabel("Cantidad de personas")
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter("{x:,.0f}"))

# --- Título en dos líneas, alineado a la izquierda ---
title = ("Diferencia entre Hombres y Mujeres\n"
         "Sistema Nacional de Investigadoras e Investigadores")
ax.set_title(title, loc="left", fontsize=14)

# --- Anotaciones de la diferencia ---
# Se itera sobre cada año y se anota la diferencia (valor absoluto) entre las series
# en el centro vertical entre ambas curvas.
last_year = total_anios.index.max()
for anio, h_value in total_anios["H"].items():
    if anio % 5 == 0 or anio == last_year:
        m_value = total_anios["M"].loc[anio]
        diff = abs(h_value - m_value)
        mid_y = (h_value + m_value) / 2
        ax.annotate(f"{diff:,.0f}",
                    xy=(anio, mid_y),
                    ha="center", va="center",
                    fontsize=12, color="black",
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))

# ax.legend(loc="best")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker

# Paso 1: Calculamos la diferencia entre hombres y mujeres por año
diferencia = total_anios["H"] - total_anios["M"]

# Paso 2: Calculamos la tasa de crecimiento (porcentaje de cambio)
# La fórmula es: ((valor_actual - valor_anterior) / valor_anterior) * 100
tasa_crecimiento = diferencia.pct_change() * 100

# Paso 3: Graficamos la tasa de crecimiento
fig, ax = plt.subplots(figsize=(20, 6))

ax.plot(tasa_crecimiento.index, tasa_crecimiento, color="purple", marker="o", label="Tasa de Crecimiento")

# Añadimos el formato a los ejes
ax.set_xlabel("Año")
ax.set_ylabel("Tasa de Crecimiento (%)")
ax.set_title("Tasa de Crecimiento de la Diferencia entre Hombres y Mujeres\nSistema Nacional de Investigadoras e Investigadores", loc="left", fontsize=14)

# Formateamos los valores del eje Y para que se vea como porcentaje
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter("{x:.2f}%"))

# Añadimos una línea horizontal en 0 para mejor comparación
ax.axhline(0, color="black", linewidth=1)

# Añadimos un fondo de barras horizontales para facilitar la lectura
ax.grid(True, axis='x', linestyle='-', color='lightgray', alpha=0.6)

# Anotaciones para los puntos de mayor cambio
for i, val in enumerate(tasa_crecimiento):
    if not np.isnan(val):  # Aseguramos que no sea NaN
        ax.annotate(f"{val:.2f}%", 
                    xy=(tasa_crecimiento.index[i], val), 
                    xytext=(5, 10), 
                    textcoords="offset points", 
                    ha="center", va="bottom", fontsize=10, color="black")
        
# Mostramos la leyenda
ax.legend()

# Aseguramos que todos los años estén en el eje X
ax.set_xticks(tasa_crecimiento.index)

# Rotamos las etiquetas del eje X
plt.xticks(rotation=45)

plt.savefig("/home/diana/Documents/QuartoProyects/Talescope/brecha_sni_forked/tasa.png")  # Guarda en la carpeta "images"


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker

# Paso 1: Calculamos la proporción de mujeres con respecto al total (hombres + mujeres) por año
proporcion_mujeres = total_anios["M"] / (total_anios["H"] + total_anios["M"])

# Paso 2: Calculamos la tasa de crecimiento de la proporción de mujeres
# La fórmula es: ((proporcion_actual - proporcion_anterior) / proporcion_anterior) * 100
tasa_crecimiento_mujeres = proporcion_mujeres.pct_change() * 100

# Paso 3: Graficamos la tasa de crecimiento
fig, ax = plt.subplots(figsize=(15, 6))

ax.plot(tasa_crecimiento_mujeres.index, tasa_crecimiento_mujeres, color="purple", marker="o", label="Tasa de Crecimiento Proporción Mujeres")

# Añadimos el formato a los ejes
ax.set_xlabel("Año")
ax.set_ylabel("Tasa de Crecimiento (%)")
ax.set_title("Tasa de Crecimiento de la Proporción de Mujeres en el SNI\nSistema Nacional de Investigadoras e Investigadores", loc="left", fontsize=14)

# Formateamos los valores del eje Y para que se vea como porcentaje
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter("{x:.2f}%"))

# Añadimos una línea horizontal en 0 para mejor comparación
ax.axhline(0, color="black", linewidth=1)

# Añadimos un fondo de barras verticales para facilitar la lectura
ax.grid(True, axis='x', linestyle='--', color='lightgray', alpha=0.6)

# Anotaciones para los puntos de mayor cambio
for i, val in enumerate(tasa_crecimiento_mujeres):
    if not np.isnan(val):  # Aseguramos que no sea NaN
        ax.annotate(f"{val:.2f}%", 
                    xy=(tasa_crecimiento_mujeres.index[i], val), 
                    xytext=(5, 10), 
                    textcoords="offset points", 
                    ha="center", va="bottom", fontsize=10, color="black")

# Mostramos la leyenda
ax.legend()

# Aseguramos que todos los años estén en el eje X
ax.set_xticks(tasa_crecimiento_mujeres.index)

# Rotamos las etiquetas del eje X
plt.xticks(rotation=45)
plt.savefig("/home/diana/Documents/QuartoProyects/Talescope/brecha_sni_forked/tasamujeres.png")

plt.show()
```
